```python
#Shopping list app
# Create an empty list
shopping_list = []

#Reading a file to take input list
with open("shopping_list.txt", "r") as file:
  shopping_list = file.read().splitlines()

# Create a list menu
def menu():
  print("----Shopping list menu:---")
  print("1.View the list")
  print("2.Add item to the list")
  print("3.Remove item from the list")
  print("4.Edit a item from the list")
  print("5.clear the list")
  print("6.Exit")

# View of list
def view_list():
  print("----Shopping list:---")
  for item in shopping_list:
    print(f"- {item}")
  if not shopping_list:
    print("The list is empty.\n")

# Adding an item to a list
def add_item():
  item = input("Enter an item to add to the list: ")
  shopping_list.append(item)
  print(f"{item} has been added to the list.\n\n")
# Removing an item in a list
def remove_item():
  item = input("Enter an item to remove from the list: ")
  if item in shopping_list:
    shopping_list.remove(item)
    print(f"{item} has been removed from the list.\n\n")

# Editing an item in the list
def edit_item():
  item = input("Enter an item to edit: ")
  if item in shopping_list:
    new_item = input("Enter the new item: ")
    index = shopping_list.index(item)
    shopping_list[index] = new_item
    print(f"{item} has been edited to {new_item} in the list.\n\n")
  else:
    print(f"{item} is not in the list.\n\n")

# Clear the list
def clear_list():
  shopping_list.clear()
  print("The list has been cleared.\n\n")

#Asking user option/choice
def User_input():
  while(True):
    menu()
    user_input = input("Enter your choice: ")
    if user_input == "1":
      view_list()
    elif user_input == "2":
      add_item()
    elif user_input == "3":
      remove_item()
    elif user_input == "4":
      edit_item()
    elif user_input == "5":
      clear_list()
    elif user_input == "6":
      print("Hey your list is ready.")
      with open("shopping_list.txt", "w") as file:
        for item in shopping_list:
          file.write(item + "\n")
        print("Happy shopping!!!")
      break
#start listing
User_input()

```
