```python
#Contact Book
#contact
contacts={}

#Add contact
def add_contact():

    print("________________________________________________________________________________________\n")
    print("Enter your contact details:")
    name=input("Name:")
    phoneNumber=list(map(int,input("Phone number:").split(",")))
    email=list(map(str,input("Email:").split(",")))
    contacts[name]={'phoneNumber':phoneNumber,'email':email}
    print("Your contact is added")
    print("________________________________________________________________________________________\n")
#del contact
def del_contact(name):
    print("________________________________________________________________________________________\n")
    if name in contacts:
        del contacts[name]
        print("Your contact is deleted")
    else:
        print("Your contact is not found")
    print("________________________________________________________________________________________\n")
#view all contacts
def all_contacts():
  print("________________________________________________________________________________________\n")
  if len(contacts)==0:
    print("\n\nYour contact book is empty")
  else:
    print("\n\n__________________________________________________________________________________________")
    print("All contacts")
    print("__________________________________________________________________________________________")
    i=0
    for name in contacts:
      i+=1
      print(f"{i}. Name:{name}\nPhone:{contacts[name]['phoneNumber']}\nEmail:{contacts[name]['email']}\n")

    print(f"{i} contacts are available")
  print("________________________________________________________________________________________\n")

#edit contact by name
def edit_contact(name):
  if name not in contacts:
    print("Your contact is not found")
  else:
    print(f"Name:{name}\nPhone:{contacts[name]['phoneNumber']}\nEmail:{contacts[name]['email']}")
    print("________________________________________________________________________________________\n")
    print("What do you want to edit:")
    x=input("1.Name\n2.Phone\n3.Email\n")
    if x=='1':
      new_name=input("Enter new name:")
      contacts[new_name]=contacts.pop(name)
      print("Your contact is edited")
    elif x=='2':
      new_phone=input("Enter new phone:")
      contacts[name]['phoneNumber']=new_phone
      print("Your contact is edited")
    elif x=='3':
      new_email=input("Enter new email:")
      contacts[name]['email']=new_email
      print("Your contact is edited")
    else:
      print("Invalid input")
  print("________________________________________________________________________________________\n")

#Search contact
def search_contact(name):
  print("\n\n__________________________________________________\nSearching.......\n___________________________________________________________")
  if name not in contacts:
    print("Your contact is not found")
  else:
    print(f"Name:{name}\nPhone:{contacts[name]['phoneNumber']}\nEmail:{contacts[name]['email']}")
#delete a contact
def delete_contact(name):
  if name not in contacts:
    print("Your contact is not found")
  else:
    print("\n_______________________________________________\nDeleting.......\n______________________________________________________________")
    del contacts[name]
    print("Your contact is deleted")

  print("________________________________________________________________________________________________\n")

#user options
def user_options():
  print("________________________________________________________________________________________\n")
  print("What do you want to do?")
  print("1.Add contact\n2.Delete contact\n3.Edit contact\n4.Search contact\n5.View all contacts\n6.Exit")
  print("________________________________________________________________________________________\n")

#start the app
def start():
  print("Welcome to Contact Book App\n your useful contacts are with us")
  print("________________________________________________________________________________________\n")
  #get user input
  while(True):
    user_options()
    x=input("Enter your choice:")
    if x=='1':
      add_contact()
    elif x=='2':
      name=input("Enter name of contact to delete:")
      delete_contact(name)
    elif x=='3':
      name=input("Enter name of contact to edit:")
      edit_contact(name)
    elif x=='4':
      name=input("Enter name of contact to search:")
      search_contact(name)
    elif x=='5':
      all_contacts()
    elif x=='6':
      print("Thank you for using our app")
      break
    else:
      print("Invalid input")
start()
```
