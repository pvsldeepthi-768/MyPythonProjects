```python
#Creating Math Quiz
import random
import time

#function for generating question
def generate_question(upper ,lower):
    num1 = random.randint(lower, upper)
    num2 = random.randint(lower, upper)
    operator = random.choice(['+', '-', '*','/','%'])
    if operator == '+':
        answer = num1 + num2
    elif operator == '-':
        answer = num1 - num2
    elif operator == '*':
        answer = num1 * num2
    elif operator == '/':
        answer = round((num1 / num2),2)
    elif operator == "%":
        answer= num1 % num2
    if(operator=='/'):
        return f"{num1} {operator} {num2}(in 2 decimals)",round(answer,2)
    return f"{num1} {operator} {num2}",answer
# Math Quiz Process
def Math_Quiz(time_limit,upper,lower):
  print("\n\n-------------Your quiz gonna starts now------------\n")
  score=0
  #ask no of questions
  n=int(input("Enter no of questions: "))
  for i in range(n):
    question,answer=generate_question(upper,lower)
    print(f"\n\nQuestion {i+1}: {question}")
    print("You have 15 sec time to answer")
    #set 15 sec timer to get answer from user
    start_time=time.time()
    user_answer=float(input("Enter your answer: "))
    end_time=time.time()
    if( end_time-start_time>=time_limit):
      print("Oops you took long time to answer!!!")
      print(f"Correct Answer is {answer}\n\n")
    else:
      if(user_answer==answer):
        print("Correct Answer!!!\n\n")
        score+=1
      else:
        print(f"Oops!!! Wrong answer. \nCorrect Answer is {answer}\n\n")
  print("\n\n------------------------------------------------------\n\n")
  print("----------------Game is over!!!------------------------\n\n")
  if(score == 5):
    print(f"Great!!! you got {score}/{n}")
  elif(score//2>=n/2):
    print(f"Well Keep it up!!{score}/{n}")
  else:
    print(f"No problem! Keep practicing {score} / {n}")

#Math quiz
#Gaming levels:
print("Choose one \n Level : 1  Level : 2 Level : 3")
level=int(input("Level:"))
if(level ==1):
  upper,lower=10,1
elif(level==2):
  upper,lower=100,10
else:
  upper,lower=1000,100
print("Choose your mode:\n1. Easy\n2. Medium\n3. Hard\n4.set own time:")
mode=int(input("Mode:"))
if(mode==1):
  time_limit=30
elif(mode==2):
  time_limit=15
elif(mode==3):
  time_limit=10
else:
  time_limit=int(input("Enter time limit:"))
Math_Quiz(time_limit,upper,lower)

```
